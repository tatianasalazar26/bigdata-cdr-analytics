### 🥉 CAPA BRONZE: Datos Crudos

**Objetivo:** Ingestar datos tal como llegan, sin transformaciones, manteniendo trazabilidad completa.

**Características:**
- Datos crudos sin transformación
- Formato Delta Lake para transacciones ACID
- Columnas de auditoría (_ingested_at, _source_file)
- Schema evolution habilitado

````python
%sql
-- Crear catálogo y esquemas para arquitectura Medallion
CREATE CATALOG IF NOT EXISTS cdr_analytics;
USE CATALOG cdr_analytics;

CREATE SCHEMA IF NOT EXISTS bronze COMMENT 'Capa Bronze - Datos crudos sin procesar';
CREATE SCHEMA IF NOT EXISTS silver COMMENT 'Capa Silver - Datos limpios y enriquecidos';
CREATE SCHEMA IF NOT EXISTS gold COMMENT 'Capa Gold - Datos agregados para analytics';

SELECT 'Esquemas creados exitosamente' AS status;
````

````python
# Crear tabla Bronze con datos crudos + columnas de auditoría
from pyspark.sql.functions import current_timestamp, lit

# Añadir columnas de auditoría
df_bronze = df_cdr \
    .withColumn("_ingested_at", current_timestamp()) \
    .withColumn("_source_file", lit("CDR.csv")) \
    .withColumn("_bronze_load_date", F.current_date())

# Guardar en capa Bronze
table_bronze = "cdr_analytics.bronze.cdr_raw"

df_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .option("mergeSchema", "true") \
    .saveAsTable(table_bronze)

print(f"✅ Tabla Bronze creada: {table_bronze}")
print(f"📊 Registros en Bronze: {spark.table(table_bronze).count():,}")
````

### 🥈 CAPA SILVER: Datos Limpios y Enriquecidos

**Objetivo:** Limpiar, validar y enriquecer los datos para análisis.

**Transformaciones:**
1. **Limpieza de datos**: Eliminar duplicados, valores nulos
2. **Normalización**: Convertir tipos de datos, formatos consistentes
3. **Enriquecimiento**: Agregar información calculada
4. **Validación**: Reglas de calidad de datos

````python
# Leer desde Bronze y crear tabla Silver con transformaciones
from pyspark.sql.functions import col, when, trim, to_date, current_timestamp, round as spark_round, regexp_replace

df_silver = spark.table("cdr_analytics.bronze.cdr_raw") \
    .dropDuplicates(["call_id"]) \
    .withColumn("call_datetime", col("_Time")) \
    .withColumn("call_date", to_date(col("_Time"))) \
    .withColumn("duration_seconds", col("_Seconds").cast("integer")) \
    .withColumn("duration_minutes", col("_Minutes").cast("integer")) \
    .withColumn("cost_usd", regexp_replace(col("amount"), "\\.", "").cast("double") / 1000000) \
    .withColumn("cost_per_minute", 
                when(col("duration_minutes") > 0, 
                     spark_round(col("cost_usd") / col("duration_minutes"), 4))
                .otherwise(0)) \
    .withColumn("customer_clean", trim(col("Customer"))) \
    .withColumn("vendor_clean", trim(col("Vendor"))) \
    .withColumn("traffic_type", trim(col("Tipo_trafico"))) \
    .withColumn("country", 
                when(col("Country").isNull() | (col("Country") == "Not Applicable"), "UNKNOWN")
                .otherwise(col("Country"))) \
    .withColumn("is_zero_cost", when(col("cost_usd") == 0, True).otherwise(False)) \
    .withColumn("is_short_call", when(col("duration_seconds") < 30, True).otherwise(False)) \
    .withColumn("from_number", col("_From")) \
    .withColumn("to_number", col("_To").cast("string")) \
    .withColumn("_silver_processed_at", current_timestamp())

df_silver_final = df_silver.select(
    "call_id", "call_datetime", "call_date", "from_number", "to_number",
    "duration_seconds", "duration_minutes", "cost_usd", "cost_per_minute",
    "customer_clean", "vendor_clean", "traffic_type", "country", "_Description",
    "is_zero_cost", "is_short_call", "_silver_processed_at", "_ingested_at"
)

# Guardar en capa Silver
table_silver = "cdr_analytics.silver.cdr_cleaned"
df_silver_final.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(table_silver)

print(f"✅ Tabla Silver creada: {table_silver}")
print(f"📊 Registros en Silver: {spark.table(table_silver).count():,}")
````

````python
# Validación de calidad de datos en Silver
df_silver_check = spark.table("cdr_analytics.silver.cdr_cleaned")

print("🔍 VALIDACIÓN DE CALIDAD - CAPA SILVER")
print("=" * 60)

# Métricas de calidad
quality_metrics = df_silver_check.select(
    F.count("*").alias("total_records"),
    F.sum(when(col("call_datetime").isNull(), 1).otherwise(0)).alias("null_datetime"),
    F.sum(when(col("duration_seconds") < 0, 1).otherwise(0)).alias("negative_duration"),
    F.sum(when(col("cost_usd") < 0, 1).otherwise(0)).alias("negative_cost"),
    F.sum(when(col("is_zero_cost"), 1).otherwise(0)).alias("zero_cost_calls"),
    F.sum(when(col("is_short_call"), 1).otherwise(0)).alias("short_calls"),
    F.countDistinct("call_id").alias("unique_calls")
)

display(quality_metrics)

print("\n✅ Validación completada")
````

### 🥇 CAPA GOLD: Tablas Analíticas

**Objetivo:** Crear tablas optimizadas y pre-agregadas listas para consumo de negocio.

**Tablas Gold:**
1. **cdr_customer_metrics** - Métricas por cliente
2. **cdr_vendor_performance** - Performance por vendor
3. **cdr_fraud_detection** - Detección de anomalías
4. **cdr_billing_summary** - Resumen para facturación

````python
# TABLA GOLD 1: Métricas por Cliente
df_customer_metrics = spark.table("cdr_analytics.silver.cdr_cleaned") \
    .groupBy("customer_clean", "call_date", "traffic_type") \
    .agg(
        F.count("*").alias("total_calls"),
        F.sum("duration_seconds").alias("total_duration_seconds"),
        F.avg("duration_seconds").alias("avg_duration_seconds"),
        F.sum("cost_usd").alias("total_revenue_usd"),
        F.avg("cost_usd").alias("avg_cost_per_call"),
        F.sum(when(col("is_zero_cost"), 1).otherwise(0)).alias("zero_cost_calls"),
        F.sum(when(col("is_short_call"), 1).otherwise(0)).alias("short_calls"),
        F.countDistinct("vendor_clean").alias("vendors_used"),
        F.countDistinct("country").alias("countries_reached"),
        F.max("_silver_processed_at").alias("last_updated")
    ) \
    .withColumn("revenue_per_minute", 
                spark_round(col("total_revenue_usd") / (col("total_duration_seconds") / 60), 4)) \
    .withColumn("zero_cost_rate", 
                spark_round(col("zero_cost_calls") / col("total_calls") * 100, 2))

table_gold_customer = "cdr_analytics.gold.cdr_customer_metrics"

df_customer_metrics.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(table_gold_customer)

print(f"✅ Tabla Gold Customer creada: {table_gold_customer}")
print(f"📊 Registros: {spark.table(table_gold_customer).count():,}")
````

````python
# TABLA GOLD 2: Performance por Vendor
df_vendor_performance = spark.table("cdr_analytics.silver.cdr_cleaned") \
    .groupBy("vendor_clean", "call_date", "traffic_type") \
    .agg(
        F.count("*").alias("total_calls"),
        F.sum("duration_seconds").alias("total_duration_seconds"),
        F.avg("duration_seconds").alias("avg_duration_seconds"),
        F.sum("cost_usd").alias("total_cost_usd"),
        F.avg("cost_usd").alias("avg_cost_per_call"),
        F.countDistinct("customer_clean").alias("customers_served"),
        F.countDistinct("country").alias("countries_covered"),
        F.sum(when(col("is_zero_cost"), 1).otherwise(0)).alias("zero_cost_calls"),
        F.max("_silver_processed_at").alias("last_updated")
    ) \
    .withColumn("cost_per_minute", 
                spark_round(col("total_cost_usd") / (col("total_duration_seconds") / 60), 4)) \
    .withColumn("avg_call_duration_minutes", 
                spark_round(col("avg_duration_seconds") / 60, 2))

table_gold_vendor = "cdr_analytics.gold.cdr_vendor_performance"

df_vendor_performance.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(table_gold_vendor)

print(f"✅ Tabla Gold Vendor creada: {table_gold_vendor}")
print(f"📊 Registros: {spark.table(table_gold_vendor).count():,}")
````

````python
# TABLA GOLD 3: Detección de Anomalías / Fraude
from pyspark.sql.window import Window

# Calcular estadísticas por cliente para detectar anomalías
window_customer = Window.partitionBy("customer_clean")

df_fraud = spark.table("cdr_analytics.silver.cdr_cleaned") \
    .withColumn("avg_duration_customer", F.avg("duration_seconds").over(window_customer)) \
    .withColumn("avg_cost_customer", F.avg("cost_usd").over(window_customer)) \
    .withColumn("stddev_duration_customer", F.stddev("duration_seconds").over(window_customer)) \
    .withColumn("stddev_cost_customer", F.stddev("cost_usd").over(window_customer))

# Detectar anomalías (valores fuera de 3 desviaciones estándar)
df_fraud_detection = df_fraud \
    .withColumn("is_duration_anomaly",
                when(
                    (col("duration_seconds") > col("avg_duration_customer") + 3 * col("stddev_duration_customer")) |
                    (col("duration_seconds") < col("avg_duration_customer") - 3 * col("stddev_duration_customer")),
                    True
                ).otherwise(False)) \
    .withColumn("is_cost_anomaly",
                when(
                    (col("cost_usd") > col("avg_cost_customer") + 3 * col("stddev_cost_customer")) |
                    (col("cost_usd") < col("avg_cost_customer") - 3 * col("stddev_cost_customer")),
                    True
                ).otherwise(False)) \
    .withColumn("is_suspicious",
                when(
                    col("is_duration_anomaly") | col("is_cost_anomaly") | 
                    col("is_zero_cost") | (col("duration_seconds") > 3600),
                    True
                ).otherwise(False)) \
    .withColumn("anomaly_score",
                (when(col("is_duration_anomaly"), 1).otherwise(0) +
                 when(col("is_cost_anomaly"), 1).otherwise(0) +
                 when(col("is_zero_cost"), 1).otherwise(0) +
                 when(col("duration_seconds") > 3600, 1).otherwise(0)))

# Filtrar solo llamadas sospechosas
df_fraud_final = df_fraud_detection \
    .filter(col("is_suspicious")) \
    .select(
        "call_id", "call_datetime", "customer_clean", "vendor_clean",
        "from_number", "to_number", "duration_seconds", "cost_usd",
        "is_duration_anomaly", "is_cost_anomaly", "is_zero_cost",
        "anomaly_score", "traffic_type", "country"
    )

table_gold_fraud = "cdr_analytics.gold.cdr_fraud_detection"

df_fraud_final.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(table_gold_fraud)

print(f"✅ Tabla Gold Fraude creada: {table_gold_fraud}")
print(f"⚠️ Llamadas sospechosas detectadas: {spark.table(table_gold_fraud).count():,}")
````

````python
# TABLA GOLD 4: Resumen de Facturación
df_billing = spark.table("cdr_analytics.silver.cdr_cleaned") \
    .groupBy("customer_clean", "vendor_clean", 
             F.year("call_date").alias("year"),
             F.month("call_date").alias("month")) \
    .agg(
        F.count("*").alias("total_calls"),
        F.sum("duration_seconds").alias("total_duration_seconds"),
        F.round(F.sum("duration_seconds") / 60, 2).alias("total_duration_minutes"),
        F.sum("cost_usd").alias("total_amount_usd"),
        F.avg("cost_usd").alias("avg_cost_per_call"),
        F.min("call_date").alias("period_start_date"),
        F.max("call_date").alias("period_end_date"),
        F.countDistinct("country").alias("countries_called"),
        F.sum(when(col("is_zero_cost"), 1).otherwise(0)).alias("free_calls"),
        F.max("_silver_processed_at").alias("last_updated")
    ) \
    .withColumn("billable_calls", col("total_calls") - col("free_calls")) \
    .withColumn("revenue_per_minute", 
                spark_round(col("total_amount_usd") / col("total_duration_minutes"), 4)) \
    .withColumn("billing_period", 
                concat_ws("-", col("year"), F.lpad(col("month"), 2, "0"))) \
    .orderBy("customer_clean", "year", "month")

table_gold_billing = "cdr_analytics.gold.cdr_billing_summary"

df_billing.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(table_gold_billing)

print(f"✅ Tabla Gold Billing creada: {table_gold_billing}")
print(f"📊 Registros de facturación: {spark.table(table_gold_billing).count():,}")
````

## 5. MODELOS DE MACHINE LEARNING Y ANALYTICS AVANZADO

### 🧠 Análisis Descriptivo y Modelado

Implementaremos tres tipos de modelos:
1. **Análisis Descriptivo**: Estadísticas y patrones de uso
2. **Predicción de Demanda**: Forecasting de llamadas
3. **Detección de Anomalías**: Identificación de patrones inusuales

````python
# ANÁLISIS DESCRIPTIVO COMPLETO
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("📊 ANÁLISIS DESCRIPTIVO COMPLETO")
print("=" * 70)

# Cargar datos desde Silver
df_analysis = spark.table("cdr_analytics.silver.cdr_cleaned").toPandas()

# Estadísticas generales
print("\n📈 ESTADÍSTICAS GENERALES:")
print(f"  - Total de llamadas: {len(df_analysis):,}")
print(f"  - Duración promedio: {df_analysis['duration_seconds'].mean():.2f} segundos")
print(f"  - Costo promedio: ${df_analysis['cost_usd'].mean():.4f}")
print(f"  - Ingreso total: ${df_analysis['cost_usd'].sum():,.2f}")
print(f"  - Clientes únicos: {df_analysis['customer_clean'].nunique()}")
print(f"  - Vendors únicos: {df_analysis['vendor_clean'].nunique()}")

# Distribución por cliente
print("\n👥 TOP 5 CLIENTES POR VOLUMEN:")
top_customers = df_analysis.groupby('customer_clean').agg({
    'call_id': 'count',
    'cost_usd': 'sum',
    'duration_seconds': 'sum'
}).rename(columns={
    'call_id': 'Total_Llamadas',
    'cost_usd': 'Ingreso_Total',
    'duration_seconds': 'Duracion_Total'
}).sort_values('Ingreso_Total', ascending=False).head(5)

print(top_customers.to_string())

# Distribución por vendor
print("\n🏭 TOP 5 VENDORS POR COSTO:")
top_vendors = df_analysis.groupby('vendor_clean').agg({
    'call_id': 'count',
    'cost_usd': 'sum'
}).rename(columns={
    'call_id': 'Total_Llamadas',
    'cost_usd': 'Costo_Total'
}).sort_values('Costo_Total', ascending=False).head(5)

print(top_vendors.to_string())

print("\n✅ Análisis descriptivo completado")
````

````python
# MODELO 1: PREDICCIÓN DE DEMANDA (Forecasting con Prophet)
print("🔮 MODELO DE PREDICCIÓN DE DEMANDA")
print("=" * 70)

# Instalar Prophet si no está disponible
%pip install prophet --quiet

from prophet import Prophet
import pandas as pd

# Preparar datos para Prophet (agregación diaria)
df_daily = spark.sql("""
    SELECT 
        call_date as ds,
        COUNT(*) as y
    FROM cdr_analytics.silver.cdr_cleaned
    GROUP BY call_date
    ORDER BY call_date
""").toPandas()

print(f"\n📅 Datos de entrenamiento: {len(df_daily)} días")
print(f"   Rango: {df_daily['ds'].min()} a {df_daily['ds'].max()}")

# Validar datos suficientes para Prophet (mínimo 2 días)
if len(df_daily) < 2:
    print("\n⚠️ ADVERTENCIA: Datos insuficientes para modelado")
    print(f"   - Prophet requiere al menos 2 días de datos históricos")
    print(f"   - Disponibles: {len(df_daily)} día(s)")
    print(f"\n💡 RECOMENDACIÓN:")
    print("   - Cargar datos históricos de múltiples fechas")
    print("   - El dataset actual contiene solo registros del {}".format(df_daily['ds'].iloc[0]))
    print("   - Para forecasting efectivo, se requieren al menos 2 semanas de datos")
    print("\n⏭️ Saltando entrenamiento de modelo Prophet")
else:
    # Entrenar modelo Prophet
    model = Prophet(
        daily_seasonality=True,
        yearly_seasonality=True,
        weekly_seasonality=True,
        seasonality_mode='multiplicative'
    )
    
    model.fit(df_daily)
    print("\n✅ Modelo entrenado exitosamente")
    
    # Hacer predicciones para los próximos 30 días
    future = model.make_future_dataframe(periods=30)
    forecast = model.predict(future)
    
    # Mostrar predicciones futuras
    print("\n🔮 PREDICCIONES PARA LOS PRÓXIMOS 7 DÍAS:")
    future_predictions = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(7)
    future_predictions.columns = ['Fecha', 'Predicción', 'Límite_Inferior', 'Límite_Superior']
    print(future_predictions.to_string(index=False))
    
    # Gráfico de predicción
    fig = model.plot(forecast, figsize=(12, 6))
    plt.title('Predicción de Demanda de Llamadas (30 días)', fontsize=14, fontweight='bold')
    plt.xlabel('Fecha')
    plt.ylabel('Número de Llamadas')
    plt.tight_layout()
    plt.show()
    
    print("\n✅ Modelo de predicción completado")
````

````python
# MODELO 2: SEGMENTACIÓN DE CLIENTES (K-Means Clustering)
print("🎯 MODELO DE SEGMENTACIÓN DE CLIENTES")
print("=" * 70)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Preparar datos de clientes
df_customer_features = spark.sql("""
    SELECT 
        customer_clean,
        COUNT(*) as total_calls,
        SUM(cost_usd) as total_revenue,
        AVG(duration_seconds) as avg_duration,
        SUM(CASE WHEN is_zero_cost THEN 1 ELSE 0 END) as zero_cost_calls,
        COUNT(DISTINCT vendor_clean) as vendors_used,
        COUNT(DISTINCT country) as countries_called
    FROM cdr_analytics.silver.cdr_cleaned
    GROUP BY customer_clean
""").toPandas()

print(f"\n👥 Clientes a segmentar: {len(df_customer_features)}")

# Preparar features para clustering
features = ['total_calls', 'total_revenue', 'avg_duration', 'zero_cost_calls', 'vendors_used', 'countries_called']
X = df_customer_features[features]

# Normalizar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar K-Means (3 segmentos)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_customer_features['segment'] = kmeans.fit_predict(X_scaled)

# Etiquetar segmentos
segment_labels = {
    0: 'Premium',
    1: 'Standar',
    2: 'Basic'
}

# Asignar labels basados en revenue promedio por segmento
segment_revenue = df_customer_features.groupby('segment')['total_revenue'].mean().sort_values(ascending=False)
for i, (segment, _) in enumerate(segment_revenue.items()):
    segment_labels[segment] = ['Premium', 'Standard', 'Basic'][i]

df_customer_features['segment_label'] = df_customer_features['segment'].map(segment_labels)

# Resumen de segmentos
print("\n🏷️ SEGMENTOS DE CLIENTES:")
segment_summary = df_customer_features.groupby('segment_label').agg({
    'customer_clean': 'count',
    'total_calls': 'mean',
    'total_revenue': 'mean',
    'avg_duration': 'mean'
}).round(2)

segment_summary.columns = ['Clientes', 'Llamadas_Promedio', 'Revenue_Promedio', 'Duracion_Promedio']
print(segment_summary.to_string())

# Guardar segmentación en Gold
df_segments_spark = spark.createDataFrame(df_customer_features[['customer_clean', 'segment_label', 'total_calls', 'total_revenue']])
df_segments_spark.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("cdr_analytics.gold.customer_segments")

print("\n✅ Segmentación de clientes completada")
print("💾 Resultados guardados en: cdr_analytics.gold.customer_segments")
````

````python
# MODELO 3: SISTEMA DE DETECCIÓN DE ANOMALÍAS AVANZADO
print("⚠️ SISTEMA DE DETECCIÓN DE ANOMALÍAS")
print("=" * 70)

from sklearn.ensemble import IsolationForest

# Cargar datos para detección de anomalías
df_anomaly = spark.sql("""
    SELECT 
        call_id,
        customer_clean,
        duration_seconds,
        cost_usd,
        HOUR(call_datetime) as call_hour,
        CASE WHEN is_zero_cost THEN 1 ELSE 0 END as is_zero_cost
    FROM cdr_analytics.silver.cdr_cleaned
""").toPandas()

print(f"\n🔍 Analizando {len(df_anomaly):,} llamadas...")

# Preparar features
X_anomaly = df_anomaly[['duration_seconds', 'cost_usd', 'call_hour', 'is_zero_cost']]

# Entrenar Isolation Forest
iso_forest = IsolationForest(
    contamination=0.05,  # Esperamos 5% de anomalías
    random_state=42,
    n_estimators=100
)

df_anomaly['anomaly_prediction'] = iso_forest.fit_predict(X_anomaly)
df_anomaly['anomaly_score'] = iso_forest.score_samples(X_anomaly)

# -1 indica anomalía, 1 indica normal
df_anomaly['is_anomaly'] = df_anomaly['anomaly_prediction'] == -1

anomalies_detected = df_anomaly['is_anomaly'].sum()
print(f"\n⚠️ Anomalías detectadas: {anomalies_detected:,} ({anomalies_detected/len(df_anomaly)*100:.2f}%)")

# Top anomalías (scores más bajos)
print("\n🔴 TOP 10 LLAMADAS MÁS SOSPECHOSAS:")
top_anomalies = df_anomaly[df_anomaly['is_anomaly']].nsmallest(10, 'anomaly_score')[[
    'call_id', 'customer_clean', 'duration_seconds', 'cost_usd', 'anomaly_score'
]]
print(top_anomalies.to_string(index=False))

# Guardar anomalías detectadas
df_anomalies_spark = spark.createDataFrame(
    df_anomaly[df_anomaly['is_anomaly']][['call_id', 'customer_clean', 'duration_seconds', 'cost_usd', 'anomaly_score']]
)

df_anomalies_spark.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("cdr_analytics.gold.ml_anomaly_detection")

print("\n✅ Sistema de detección de anomalías completado")
print("💾 Resultados guardados en: cdr_analytics.gold.ml_anomaly_detection")
````

````python
# VISUALIZACIONES PROFESIONALES CON PLOTLY
print("📊 GENERANDO DASHBOARDS INTERACTIVOS PROFESIONALES...\n")

%pip install plotly kaleido --quiet

import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd

# Configuración de tema corporativo
COLOR_PALETTE = {
    'primary': '#1f77b4',
    'success': '#2ca02c',
    'warning': '#ff7f0e',
    'danger': '#d62728',
    'info': '#17becf',
    'premium': '#FFD700',
    'standard': '#C0C0C0',
    'basic': '#CD7F32'
}

# =============================================================================
# 1. DASHBOARD - KPIs PRINCIPALES
# =============================================================================
print("📈 1. Generando Dashboard de KPIs...")

from plotly.subplots import make_subplots
import plotly.graph_objects as go

# =========================
# CONSULTA DE KPIs
# =========================

df_kpis = spark.sql("""
    SELECT
        COUNT(*) as total_llamadas,
        ROUND(SUM(cost_usd), 2) as revenue_total,
        COUNT(DISTINCT customer_clean) as clientes_activos,
        ROUND(SUM(duration_seconds) / 3600, 0) as horas_totales,
        COUNT(DISTINCT vendor_clean) as vendors_activos
    FROM cdr_analytics.silver.cdr_cleaned
""").toPandas().iloc[0]

df_anomalias = spark.sql("""
    SELECT COUNT(*) as anomalias
    FROM cdr_analytics.gold.cdr_fraud_detection
""").toPandas().iloc[0]["anomalias"]

# =========================
# DEFINICIÓN DE KPIs
# =========================

kpis = [
    {
        "label": "Total Llamadas",
        "value": int(df_kpis["total_llamadas"]),
        "icon": "📞",
        "color": COLOR_PALETTE["primary"],
        "format": ","
    },
    {
        "label": "Revenue Total (USD)",
        "value": float(df_kpis["revenue_total"]),
        "icon": "💰",
        "color": COLOR_PALETTE["success"],
        "format": "$,.2f"
    },
    {
        "label": "Clientes Activos",
        "value": int(df_kpis["clientes_activos"]),
        "icon": "👥",
        "color": COLOR_PALETTE["info"],
        "format": ","
    },
    {
        "label": "Horas Procesadas",
        "value": int(df_kpis["horas_totales"]),
        "icon": "⏱️",
        "color": COLOR_PALETTE["warning"],
        "format": ","
    },
    {
        "label": "Vendors Activos",
        "value": int(df_kpis["vendors_activos"]),
        "icon": "🏢",
        "color": COLOR_PALETTE["primary"],
        "format": ","
    },
    {
        "label": "Anomalías Detectadas",
        "value": int(df_anomalias),
        "icon": "⚠️",
        "color": COLOR_PALETTE["danger"],
        "format": ","
    }
]

# =========================
# DASHBOARD
# =========================

fig_kpis = make_subplots(
    rows=2,
    cols=3,
    subplot_titles=[
        f"{k['icon']} {k['label']}"
        for k in kpis
    ],
    specs=[
        [{"type": "indicator"}] * 3,
        [{"type": "indicator"}] * 3
    ]
)

for i, kpi in enumerate(kpis):

    row = (i // 3) + 1
    col = (i % 3) + 1

    fig_kpis.add_trace(
        go.Indicator(
            mode="number",
            value=kpi["value"],
            number={
                "valueformat": kpi["format"],
                "font": {
                    "size": 48,
                    "color": kpi["color"]
                }
            }
        ),
        row=row,
        col=col
    )

# =========================
# ESTILO
# =========================

fig_kpis.update_layout(
    title={
        "text": "<b>📊 Dashboard - KPIs Principales</b>",
        "x": 0.5,
        "font": {"size": 26}
    },
    height=600,
    showlegend=False,
    paper_bgcolor="white",
    plot_bgcolor="#f8f9fa",
    margin=dict(
        t=100,
        l=30,
        r=30,
        b=30
    )
)

# Agrandar títulos de cada KPI
for annotation in fig_kpis['layout']['annotations']:
    annotation['font'] = dict(
        size=18,
        color='#2c3e50'
    )

fig_kpis.show()

# =============================================================================
# 2. TOP 10 CLIENTES - GRÁFICO DE BARRAS HORIZONTAL INTERACTIVO
# =============================================================================
print("👥 2. Generando Análisis de Top Clientes...")

df_top_customers = spark.sql("""
    SELECT 
        customer_clean,
        SUM(total_calls) as llamadas,
        ROUND(SUM(total_revenue_usd), 2) as revenue
    FROM cdr_analytics.gold.cdr_customer_metrics
    GROUP BY customer_clean
    ORDER BY revenue DESC
    LIMIT 10
""").toPandas()

fig_customers = go.Figure()
fig_customers.add_trace(go.Bar(
    y=df_top_customers['customer_clean'][::-1],
    x=df_top_customers['revenue'][::-1],
    orientation='h',
    marker=dict(
        color=df_top_customers['revenue'][::-1],
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title="Revenue USD")
    ),
    text=[f"${r:,.0f}" for r in df_top_customers['revenue'][::-1]],
    textposition='auto',
    hovertemplate='<b>%{y}</b><br>Revenue: $%{x:,.2f}<br>Llamadas: %{customdata:,}<extra></extra>',
    customdata=df_top_customers['llamadas'][::-1]
))

fig_customers.update_layout(
    title="<b>🏆 Top 10 Clientes por Revenue</b>",
    title_font_size=20,
    xaxis_title="Revenue (USD)",
    yaxis_title="Cliente",
    height=500,
    template='plotly_white',
    hovermode='closest'
)
fig_customers.show()

# =============================================================================
# 3. SEGMENTACIÓN DE CLIENTES - GRÁFICO DE DONA INTERACTIVO
# =============================================================================
print("🎯 3. Generando Análisis de Segmentación...")

df_segments = spark.sql("""
    SELECT 
        segment_label,
        COUNT(*) as clientes,
        ROUND(AVG(total_revenue), 2) as revenue_promedio
    FROM cdr_analytics.gold.customer_segments
    GROUP BY segment_label
    ORDER BY revenue_promedio DESC
""").toPandas()

colors_segments = [COLOR_PALETTE['premium'], COLOR_PALETTE['standard'], COLOR_PALETTE['basic']]

fig_segments = go.Figure(data=[go.Pie(
    labels=df_segments['segment_label'],
    values=df_segments['clientes'],
    hole=0.4,
    marker=dict(colors=colors_segments, line=dict(color='white', width=2)),
    textinfo='label+percent',
    textfont_size=14,
    hovertemplate='<b>%{label}</b><br>Clientes: %{value}<br>Porcentaje: %{percent}<br>Revenue Prom: $%{customdata:,.2f}<extra></extra>',
    customdata=df_segments['revenue_promedio']
)])

fig_segments.update_layout(
    title="<b>🎯 Distribución de Segmentos de Clientes</b>",
    title_font_size=20,
    height=500,
    template='plotly_white',
    annotations=[dict(text='Segmentos', x=0.5, y=0.5, font_size=20, showarrow=False)]
)
fig_segments.show()

# =============================================================================
# 4. ANÁLISIS DE VENDORS - TREEMAP INTERACTIVO
# =============================================================================
print("🏢 4. Generando Análisis de Performance de Vendors...")

df_vendors = spark.sql("""
    SELECT 
        vendor_clean,
        SUM(total_calls) as llamadas,
        ROUND(SUM(total_cost_usd), 2) as costo,
        ROUND(AVG(cost_per_minute), 4) as costo_minuto
    FROM cdr_analytics.gold.cdr_vendor_performance
    GROUP BY vendor_clean
    ORDER BY costo DESC
    LIMIT 15
""").toPandas()

fig_vendors = go.Figure(go.Treemap(
    labels=df_vendors['vendor_clean'],
    parents=[''] * len(df_vendors),
    values=df_vendors['costo'],
    text=[f"${c:,.0f}" for c in df_vendors['costo']],
    textposition='middle center',
    marker=dict(
        colorscale='RdYlGn_r',
        cmid=df_vendors['costo_minuto'].median(),
        colorbar=dict(title="Costo/min")
    ),
    hovertemplate='<b>%{label}</b><br>Costo Total: $%{value:,.2f}<br>Llamadas: %{customdata:,}<extra></extra>',
    customdata=df_vendors['llamadas']
))

fig_vendors.update_layout(
    title="<b>🏢 Distribución de Costos por Vendor (Treemap)</b>",
    title_font_size=20,
    height=600
)
fig_vendors.show()

# =============================================================================
# 5. DETECCIÓN DE ANOMALÍAS - SCATTER PLOT 3D
# =============================================================================
print("⚠️ 5. Generando Visualización de Anomalías...")

df_anomalies_viz = spark.sql("""
    SELECT 
        customer_clean,
        traffic_type,
        COUNT(*) as anomalias,
        MAX(anomaly_score) as max_score
    FROM cdr_analytics.gold.cdr_fraud_detection
    GROUP BY customer_clean, traffic_type
    ORDER BY anomalias DESC
    LIMIT 50
""").toPandas()

fig_anomalies = px.scatter(
    df_anomalies_viz,
    x='customer_clean',
    y='anomalias',
    size='anomalias',
    color='max_score',
    hover_data=['traffic_type'],
    color_continuous_scale='Reds',
    title="<b>⚠️ Detección de Anomalías por Cliente</b>",
    labels={'anomalias': 'Número de Anomalías', 'customer_clean': 'Cliente', 'max_score': 'Score Máximo'}
)

fig_anomalies.update_layout(
    height=600,
    template='plotly_white',
    xaxis={'tickangle': -45}
)
fig_anomalies.show()

# =============================================================================
# 6. DISTRIBUCIÓN DE DURACIÓN - HISTOGRAMA INTERACTIVO
# =============================================================================
print("📊 6. Generando Análisis de Distribución de Duración...")

df_duration = spark.sql("""
    SELECT duration_seconds
    FROM cdr_analytics.silver.cdr_cleaned
    WHERE duration_seconds < 600
""").toPandas()

fig_duration = go.Figure()
fig_duration.add_trace(go.Histogram(
    x=df_duration['duration_seconds'],
    nbinsx=50,
    marker=dict(
        color=COLOR_PALETTE['info'],
        line=dict(color='white', width=1)
    ),
    hovertemplate='Duración: %{x}s<br>Frecuencia: %{y}<extra></extra>'
))

fig_duration.update_layout(
    title="<b>📊 Distribución de Duración de Llamadas</b>",
    title_font_size=20,
    xaxis_title="Duración (segundos)",
    yaxis_title="Frecuencia",
    height=500,
    template='plotly_white'
)
fig_duration.show()

print("\n✅ DASHBOARDS INTERACTIVOS GENERADOS EXITOSAMENTE")
print("📊 7 visualizaciones profesionales con Plotly")
print("🎨 Diseño corporativo, interactivo y responsive")
print("💡 Listas para exportar o integrar en dashboards de BI")
````

**Tareas:**

1. **Task 1: Ingesta Bronze** 
   - Frecuencia: Cada hora
   - Acción: Leer nuevos archivos CDR y cargar a capa Bronze
   - Timeout: 30 minutos

2. **Task 2: Transformación Silver** (depende de Task 1)
   - Acción: Limpiar y enriquecer datos en capa Silver
   - Timeout: 45 minutos

3. **Task 3: Agregación Gold** (depende de Task 2)
   - Acción: Crear tablas analíticas en capa Gold
   - Timeout: 30 minutos

4. **Task 4: Modelos ML** (depende de Task 3)
   - Frecuencia: Diaria
   - Acción: Re-entrenar modelos predictivos
   - Timeout: 60 minutos

5. **Task 5: Alertas** (depende de Task 3)
   - Acción: Enviar notificaciones de anomalías detectadas
   - Webhook/Email: alertas@empresa.com
